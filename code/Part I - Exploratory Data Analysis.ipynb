## Exploratory Data Analysis

### Breast Cancer Wisconsin (Diagnostic) Data Set

https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data

```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

# Apply the default theme to seaborn
sns.set_theme()
```

```python
# Loading the dataset as a pandas df
data = pd.read_csv("../data/breast_cancer.csv")
data.shape
```

```python
# Information about the dataset
data.info()
```

```python
# Visualize the first 5 rows
data.head()
```

```python
# Counting missing values per column
data.isna().sum()
```

```python
# Drop column with NaN values
data.drop(columns = ["id", "Unnamed: 32"], inplace=True)
```

```python
# Encoding the target variable 
data = pd.get_dummies(data, columns=["diagnosis"], drop_first=True)

# Target variable "diagnosis_M" moved to the first column
data.insert(0, 'diagnosis_M', data.pop('diagnosis_M'))

# Visualizing the first 5 rows
data.head()
```

```python
# Summary statistics of the inputs
data.describe()
```

### Data Visualization

```python
# plot scatter matrix using seaborn
sns.pairplot(data.iloc[:, 0:11], hue="diagnosis_M")
plt.show()
```

```python
# violin plots
plt.violinplot(data.iloc[:, 1:11])
plt.show()
```

### Exploring 'smoothness_mean' and 'fractal_dimension_mean' 

```python
sns.scatterplot(data, x='smoothness_mean', y='fractal_dimension_mean', hue='diagnosis_M')
plt.show()
```

```python
sns.lmplot(data, x='smoothness_mean', y='fractal_dimension_mean', hue='diagnosis_M')
plt.show()
```

```python
# Violin Plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4.5), tight_layout=True)
fig.suptitle('Violin Plots')

# Smoothness Mean
ax1.set_title('Smoothness Mean')
ax1.violinplot(data['smoothness_mean'])

# Fractal Dimension Mean
ax2.set_title('Fractal Dimension Mean')
ax2.violinplot(data['fractal_dimension_mean'])

plt.show()
```

```python
# Denisty Plots
fig, axes = plt.subplots(1, 2, figsize=(9, 4.5), tight_layout=True)
fig.suptitle('Density Plots')

# Smoothness Mean
sns.kdeplot(ax=axes[0], data=data, x='smoothness_mean', hue='diagnosis_M', fill=True)
axes[0].set_title('Smoothness Mean')

# Fractal Dimension Mean
sns.kdeplot(ax=axes[1], data=data, x='fractal_dimension_mean', hue='diagnosis_M', fill=True)
axes[1].set_title('Fractal Dimension Mean')

plt.show()
```

```python
# Creating a new csv with the preprocessed data
data.to_csv("./data/breast_cancer_mod.csv", index=False)
```
